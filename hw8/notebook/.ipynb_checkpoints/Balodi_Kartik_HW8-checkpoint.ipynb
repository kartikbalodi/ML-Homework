```python
df1 = pd.read_csv('../data/wdbc.csv', header=None)
df1 = df1.drop([0], axis=1) #remove ID number since it's irrelevant to our analysis

#replace malignant/benign with 1/0 for encoding of class type
df1[1] = df1[1].replace(['M'], '1')
df1[1] = df1[1].replace(['B'], '0')
df1[1] = df1[1].astype(int)

df1_m = df1.loc[df1[1]==1] # positive class instances (malignant)
df1_b = df1.loc[df1[1]==0] # negative class instances (benign)
```

**i. Supervised Learning: Train an L1-penalized SVM to classify the data. Use 5 fold cross validation to choose the penalty parameter. Use normalized data. Report the average accuracy, precision, recall, F1-score, and AUC, for both training and test sets over your M runs. Plot the ROC and report the confusion matrix for training and testing in one of the runs.**  
  
Resources consulted:
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html
- https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.Normalizer.html
- https://stackoverflow.com/questions/49444262/normalize-data-before-or-after-split-of-training-and-testing-data

```python
accuracy_train = []
precision_train = []
recall_train = []
f1_score_train = []
auc_train = []

accuracy_test = []
precision_test = []
recall_test = []
f1_score_test = []
auc_test = []

for i in range(30):
    
    #randomly sample rows to create raw test/train sets
    df1_test = pd.concat([df1_m.sample(int(len(df1_m.index)*.2)), df1_b.sample(int(len(df1_b.index)*.2))], axis=0)
    df1_train = df1.drop(df1_test.index)
    X_test = df1_test.iloc[:,1:]
    y_test = df1_test[1]
    X_train = df1_train.iloc[:,1:]
    y_train = df1_train[1]
    
    #normalize and standardize test/train data
    normalizer = Normalizer().fit(X_train)
    temp_X_test = normalizer.transform(X_test)
    temp_X_train = normalizer.transform(X_train)
    scaler = StandardScaler().fit(temp_X_train)
    X_test_scaled = scaler.transform(temp_X_test)
    X_train_scaled = scaler.transform(temp_X_train)
    
    #initialize classifier with 5-fold CV using GridSearchCV
    svm_l1 = LinearSVC(penalty="l1",dual=False)
    reg_pen_dict = {'C':[10 ** i for i in range(-3,6)]}
    svm_l1_cv = GridSearchCV(estimator=svm_l1, param_grid=reg_pen_dict, cv=StratifiedKFold(5))
    clf = svm_l1_cv.fit(X_train_scaled,y_train)
    y_train_pred = clf.predict(X_train_scaled)
    y_test_pred = clf.predict(X_test_scaled)
    
    #report metrics
    accuracy_train.append(accuracy_score(y_train,y_train_pred)) #accuracy
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
    precision_train.append(precision_score(y_train,y_train_pred)) #precision
    precision_test.append(precision_score(y_test,y_test_pred))
    recall_train.append(recall_score(y_train,y_train_pred)) #recall
    recall_test.append(recall_score(y_test,y_test_pred))
    f1_score_train.append(f1_score(y_train,y_train_pred)) #f1-score
    f1_score_test.append(f1_score(y_test,y_test_pred))
    
    fpr,tpr,thres = roc_curve(y_train,y_train_pred) #auc
    auc_train.append(auc(fpr,tpr))
    fpr,tpr,thres = roc_curve(y_test,y_test_pred)
    auc_test.append(auc(fpr,tpr))
    
    #Plot the ROC and report the confusion matrix for training and testing in one of the runs
    if i == 0:
        fpr,tpr,thres = roc_curve(y_train,y_train_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Train Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Train Data\n',confusion_matrix(y_train,y_train_pred))
        print()
        
        fpr,tpr,thres = roc_curve(y_test,y_test_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Test Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Test Data\n',confusion_matrix(y_test,y_test_pred))
```

**ii. Semi-Supervised Learning/ Self-training: select 50% of the positive class along with 50% of the negative class in the training set as labeled data and the rest as unlabelled data. You can select them randomly.**  

Different Raw NBConvert code blocks to test various sections/instructions individually prior before I pieced them into one big code block later on

- **A. Train an L1-penalized SVM to classify the labeled data. Use normalized data. Choose the penalty parameter using 5 fold cross validation.**

- **B. Find the unlabeled data point that is the farthest to the decision boundary of the SVM. Let the SVM label it (ignore its true label), and add it to the labeled data, and retrain the SVM. Continue this process until all unlabeled data are used. Test the final SVM on the test data and the average accuracy, precision, recall, F1-score, and AUC, for both training and test sets over your M runs. Plot the ROC and report the confusion matrix for training and testing in one of the runs.**  
  
Resources consulted:
- https://stackoverflow.com/questions/32074239/sklearn-getting-distance-of-each-point-from-decision-boundary

Piecing all those individual components above together for a final form with M=30:

```python
accuracy_train = []
precision_train = []
recall_train = []
f1_score_train = []
auc_train = []

accuracy_test = []
precision_test = []
recall_test = []
f1_score_test = []
auc_test = []

for i in range(30):
    
    #randomly sample rows to create raw test/train sets
    df1_test = pd.concat([df1_m.sample(int(len(df1_m.index)*.2)), df1_b.sample(int(len(df1_b.index)*.2))], axis=0)
    df1_train = df1.drop(df1_test.index)
    X_test = df1_test.iloc[:,1:]
    y_test = df1_test[1]

    #randomly select 50% of the positive class along with 50% of the negative class in the training set as labeled data
    #and the rest as unlabeled data
    df1_train_m = df1_train.loc[df1[1]==1] # positive class instances (malignant)
    df1_train_b = df1_train.loc[df1[1]==0] # negative class instances (benign)
    df1_train_labeled = pd.concat([df1_train_m.sample(int(len(df1_train_m.index)*.5)), 
                                    df1_train_b.sample(int(len(df1_train_b.index)*.5))], axis=0)
    df1_train_unlabeled = df1_train.drop(df1_train_labeled.index)
    X_train_labeled = df1_train_labeled.iloc[:,1:]
    y_train_labeled = df1_train_labeled[1].reset_index(drop=True)
    X_train_unlabeled = df1_train_unlabeled.iloc[:,1:]

    #normalize train data (using only normalized data without standardizing since it was explicitly stated in instructions)
    normalizer = Normalizer().fit(X_train_labeled)
    X_train_labeled_normalized = normalizer.transform(X_train_labeled)
    normalizer = Normalizer().fit(X_train_unlabeled)
    X_train_unlabeled_normalized = normalizer.transform(X_train_unlabeled)

    #initialize classifier with 5-fold CV using GridSearchCV on the labelled data
    svm_l1 = LinearSVC(penalty="l1",dual=False)
    reg_pen_dict = {'C':[10 ** i for i in range(-3,6)]}
    svm_l1_cv = GridSearchCV(estimator=svm_l1, param_grid=reg_pen_dict, cv=StratifiedKFold(5))
    clf = svm_l1_cv.fit(X_train_labeled_normalized,y_train_labeled)
    y_train_pred = clf.predict(X_train_labeled_normalized)

    #add unlabelled data one by one and retrain SVM
    while(len(X_train_unlabeled_normalized)>0):
        
        relative_dist = clf.decision_function(X_train_unlabeled_normalized)
        max_index = np.argmax(np.abs(relative_dist))

        furthest_data_pt = X_train_unlabeled_normalized[max_index]
        predicted_label = clf.predict(furthest_data_pt.reshape(1,-1))

        X_train_labeled_normalized = np.append(X_train_labeled_normalized, [furthest_data_pt], axis=0)
        y_train_labeled = pd.concat([y_train_labeled,pd.DataFrame(predicted_label)],axis=0).reset_index(drop=True)

        X_train_unlabeled_normalized = np.delete(X_train_unlabeled_normalized, max_index, axis=0)

        #initialize classifier with 5-fold CV using GridSearchCV on the labelled data
        svm_l1 = LinearSVC(penalty="l1",dual=False)
        reg_pen_dict = {'C':[10 ** i for i in range(-3,6)]}
        svm_l1_cv = GridSearchCV(estimator=svm_l1, param_grid=reg_pen_dict, cv=StratifiedKFold(5))
        clf = svm_l1_cv.fit(X_train_labeled_normalized,y_train_labeled)
        y_train_pred = clf.predict(X_train_labeled_normalized)

    y_test_pred = clf.predict(Normalizer().fit_transform(X_test))

    #report metrics
    accuracy_train.append(accuracy_score(y_train_labeled,y_train_pred)) #accuracy
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
    precision_train.append(precision_score(y_train_labeled,y_train_pred)) #precision
    precision_test.append(precision_score(y_test,y_test_pred))
    recall_train.append(recall_score(y_train_labeled,y_train_pred)) #recall
    recall_test.append(recall_score(y_test,y_test_pred))
    f1_score_train.append(f1_score(y_train_labeled,y_train_pred)) #f1-score
    f1_score_test.append(f1_score(y_test,y_test_pred))

    fpr,tpr,thres = roc_curve(y_train_labeled,y_train_pred) #auc
    auc_train.append(auc(fpr,tpr))
    fpr,tpr,thres = roc_curve(y_test,y_test_pred)
    auc_test.append(auc(fpr,tpr))
    
    #Plot the ROC and report the confusion matrix for training and testing in one of the runs
    if i == 0:
        fpr,tpr,thres = roc_curve(y_train_labeled,y_train_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Train Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Train Data\n',confusion_matrix(y_train_labeled,y_train_pred))
        print()
        
        fpr,tpr,thres = roc_curve(y_test,y_test_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Test Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Test Data\n',confusion_matrix(y_test,y_test_pred))
```

**iii. Unsupervised Learning: Run k-means algorithm on the whole training set. Ignore the labels of the data, and assume k = 2.**

- **A. Run the k-means algorithm multiple times. Make sure that you initialize the algorithm randomly. How do you make sure that the algorithm was not trapped in a local minimum?**  
  
Resources consulted:
- https://stats.stackexchange.com/questions/48757/why-doesnt-k-means-give-the-global-minimum
- https://scikit-learn.org/stable/modules/clustering.html#k-means
- https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

In probabilistic terms, there is a global solution, but it requires us to iterate over all possible clusterings; the same is true if we have an objective function. This is not feasible, because say we run Kmeans clustering with k-desired clusters on n samples, we have to (n choose k) choices of centroid seeds/initial states to consider. In the case of the training data with 456 samples, this means we need to explore (456 choose 2)=103740 initial choices of starting cluster centroid.  
  
Usually, we try to avoid the algorithm from being trapped in a local minimum by setting the algorithm's 'init' argument to k-means++, which initializes the centroids to be (generally) distant from each other, leading to probably better results than random initialization. Because for this question we are specifically tasked to initialize the algorithm randomly, we need to rerun the algorithm multiple times with different cluster seeds using the 'n_init' argument (default argument is 10) and hope that one of these cluster seeds gives us a local minima that is close to/equal to the global minima. Across each Monte Carlo run, we also seed the algorithm's 'random_state' argument with different random seeds, so that different initial cluster seeds are chosen each time for the K-means algorithm to run. If the feature space is convex, the local minima will also be the global minima.

- **B. Compute the centers of the two clusters and find the closest 30 data points to each center. Read the true labels of those 30 data points and take a majority poll within them. The majority poll becomes the label predicted by k-means for the members of each cluster. Then compare the labels provided by k-means with the true labels of the training data and report the average accuracy, precision, recall, F1-score, and AUC over M runs, and ROC and the confusion matrix for one of the runs. (Here we are using k-means as a classifier. The closest 30 data points to each center are labeled by experts, so as to use k-means for classification. Obviously, this is a naive approach.)**
  
Resources consulted:
- https://scikit-learn.org/stable/modules/generated/sklearn.cluster.k_means.html
- https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.NearestNeighbors.html#sklearn.neighbors.NearestNeighbors.kneighbors
- https://stackoverflow.com/questions/34226400/find-the-index-of-the-k-smallest-values-of-a-numpy-array

- **C. Classify test data based on their proximity to the centers of the clusters. Report the average accuracy, precision, recall, F1-score, and AUC over M runs, and ROC and the confusion matrix for one of the runs for the test data. (K-means algorithm does not provide probabilities, so one can use the distances from cluster center and pass them through a softmax to calculate probabilities. Alternatively, one can calculate the ROC curve by varying the threshold for majority polling. Usually, a majority is achieved when t = 50% of the data are in a class. one can vary t and obtain an ROC curve.)**

```python
#to compare performance of raw/normalized/scaled data and later on also spectral clustering against the same random seed 
#using random_state argument
rand_list = []
for i in range(30):
    rand_list.append(np.random.randint(low=1, high=2**31-1))
```

```python
accuracy_train = []
precision_train = []
recall_train = []
f1_score_train = []
auc_train = []

accuracy_test = []
precision_test = []
recall_test = []
f1_score_test = []
auc_test = []

for i in range(30):
    
    #randomly sample rows to create raw test/train sets
    df1_test = pd.concat([df1_m.sample(int(len(df1_m.index)*.2)), df1_b.sample(int(len(df1_b.index)*.2))], axis=0)
    df1_train = df1.drop(df1_test.index)
    X_test = df1_test.iloc[:,1:]
    y_test = df1_test[1]
    X_train = df1_train.iloc[:,1:]
    y_train = df1_train[1]

    X_train = X_train.reset_index(drop=True)
    y_train = y_train.reset_index(drop=True)
    X_test = X_test.reset_index(drop=True)
    y_test = y_test.reset_index(drop=True)

    #normalize and standardize test/train data
    normalizer = Normalizer().fit(X_train)
    temp_X_test = normalizer.transform(X_test)
    temp_X_train = normalizer.transform(X_train)
    scaler = StandardScaler().fit(temp_X_train)
    X_test_scaled = scaler.transform(temp_X_test)
    X_train_scaled = scaler.transform(temp_X_train)

    #make sure you initialize the algorithm randomly
    clusterer = KMeans(n_clusters=2, init='random', n_init=100, random_state=rand_list[i])
    cluster_labels = clusterer.fit(X_train)

    cluster_distance = clusterer.transform(X_train) #returns euclidean distance between each point and cluster ([0] and [1])
    cluster_distance = np.array(cluster_distance)

    index_list = [] #store of 30 closest points by index to clusters [0] and [1]
    for j in range(cluster_distance.shape[1]):
        cluster_d = cluster_distance[:,j]
        idx = np.argpartition(cluster_d, 30)[:30]
        index_list.append(idx)

    majority_poll_cluster_label = [] #contains majority poll from true label, 1x2 list corresponding to clusters [0] and [1]
    for idx in index_list:
        true_labels = y_train.iloc[idx]
        true_cluster_counts = true_labels.value_counts()
        max_cluster_label = true_cluster_counts.sort_values(ascending=False).index[0]
        majority_poll_cluster_label.append(max_cluster_label)
    #therefore the true label of cluster [0] is majority_poll_cluster_label[0]

    #for all the training data points predicted to cluster [0], we assign it majority_poll_cluster_label[0], 
    #and likewise for cluster [1]
    y_train_pred = clusterer.predict(X_train)
    pred_labels = pd.DataFrame(y_train_pred)
    pred_labels_0 = pred_labels[pred_labels[0]==0]
    pred_labels_1 = pred_labels[pred_labels[0]==1]
    pred_labels_0['class'] = majority_poll_cluster_label[0]
    pred_labels_1['class'] = majority_poll_cluster_label[1]
    y_train_pred = pd.concat([pred_labels_0['class'],pred_labels_1['class']],axis=0).sort_index()

    #test data
    y_test_pred = clusterer.predict(X_test)
    pred_labels_test = pd.DataFrame(y_test_pred)
    pred_labels_test_0 = pred_labels_test[pred_labels_test[0]==0]
    pred_labels_test_1 = pred_labels_test[pred_labels_test[0]==1]
    pred_labels_test_0['class'] = majority_poll_cluster_label[0]
    pred_labels_test_1['class'] = majority_poll_cluster_label[1]
    y_test_pred = pd.concat([pred_labels_test_0['class'],pred_labels_test_1['class']],axis=0).sort_index()
    
    #report metrics
    accuracy_train.append(accuracy_score(y_train,y_train_pred)) #accuracy
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
    precision_train.append(precision_score(y_train,y_train_pred)) #precision
    precision_test.append(precision_score(y_test,y_test_pred))
    recall_train.append(recall_score(y_train,y_train_pred)) #recall
    recall_test.append(recall_score(y_test,y_test_pred))
    f1_score_train.append(f1_score(y_train,y_train_pred)) #f1-score
    f1_score_test.append(f1_score(y_test,y_test_pred))
    
    fpr,tpr,thres = roc_curve(y_train,y_train_pred) #auc
    auc_train.append(auc(fpr,tpr))
    fpr,tpr,thres = roc_curve(y_test,y_test_pred)
    auc_test.append(auc(fpr,tpr))
    
    #Plot the ROC and report the confusion matrix for training and testing in one of the runs
    if i == 0:
        fpr,tpr,thres = roc_curve(y_train,y_train_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Train Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Train Data\n',confusion_matrix(y_train,y_train_pred))
        print()
        
        fpr,tpr,thres = roc_curve(y_test,y_test_pred)
        plt.figure()
        plt.plot(fpr, tpr, color='green')
        plt.plot([0, 1], [0, 1], color='red', linestyle=':')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for One Run - Test Data')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.02])
        plt.show()
        plt.close()
        print('Confusion Matrix for Test Data\n',confusion_matrix(y_test,y_test_pred))
```

**(b) Repeat each of the following two procedures 50 times. You will have 50 errors for 90 SVMs per each procedure.**  
- **i. Train a SVM with a pool of 10 randomly selected data points from the training set using linear kernel and L1 penalty. Select the penalty parameter using 5-fold cross validation. Repeat this process by adding 10 other randomly selected data points to the pool, until you use all the 900 points. Do NOT replace the samples back into the training set at each step. Calculate the test error for each SVM. You will have 90 SVMs that were trained using 10, 20, 30, ... , 900 data points and their 90 test errors. You have implemented passive learning.**

```python
mc_svm_test_error = []
for i in range(50):
    
    X_test = df2_test.drop('class',axis=1)
    y_test = df2_test['class']
    passive_train_df = df2_train
    pooled_train_df = pd.DataFrame()
    
    svm_test_error = []
    run_counter = 0 
    while len(passive_train_df) > 0:
        run_counter += 1
        
        random_10_samples = passive_train_df.sample(n=10)
        if run_counter == 1: #repeat if first pool has only one class
            while(random_10_samples['class'].unique().size < 2):
                random_10_samples = passive_train_df.sample(n=10)
        
        passive_train_df = passive_train_df.drop(random_10_samples.index)
        pooled_train_df = pd.concat([pooled_train_df,random_10_samples])
        
        X_train = pooled_train_df.drop('class',axis=1)
        y_train = pooled_train_df['class']
        
        svm_l1 = LinearSVC(penalty="l1", dual=False)
        reg_pen_dict = {'C':[10 ** i for i in range(-3,6)]}
        svm_l1_cv = GridSearchCV(estimator=svm_l1, param_grid=reg_pen_dict, cv=StratifiedKFold(5))
        clf = svm_l1.fit(X_train, y_train)
        
        y_pred = clf.predict(X_test)
        svm_test_error.append(1-accuracy_score(y_test,y_pred))
    
    mc_svm_test_error.append(svm_test_error)
```

```python
mc_passive_df = pd.DataFrame(mc_svm_test_error)
```

- **ii. Train a SVM with a pool of 10 randomly selected data points from the training set (If all selected data points are from one class, select another set of 10 data points randomly) using linear kernel and L1 penalty. Select the parameters of the SVM with 5-fold cross validation. Choose the 10 closest data points in the training set to the hyperplane of the SVM (You may use the result from linear algebra about the distance of a point from a hyperplane) and add them to the pool. Do not replace the samples back into the training set. Train a new SVM using the pool. Repeat this process until all training data is used. You will have 90 SVMs that were trained using 10, 20, 30,..., 900 data points and their 90 test errors. You have implemented active learning.**

```python
mc_svm_test_error = []
for i in range(50):
    
    X_test = df2_test.drop('class',axis=1)
    y_test = df2_test['class']
    
    pooled_train_df = pd.DataFrame()
    active_train_df = df2_train
    random_10_samples = active_train_df.sample(n=10)
    while(random_10_samples['class'].unique().size < 2): #repeat if first pool has only one class
        random_10_samples = active_train_df.sample(n=10)
    
    svm_test_error = []
    while len(active_train_df) > 0:
        
        active_train_df = active_train_df.drop(random_10_samples.index)
        pooled_train_df = pd.concat([pooled_train_df,random_10_samples])
        X_train = pooled_train_df.drop('class',axis=1)
        y_train = pooled_train_df['class']
        
        svm_l1 = LinearSVC(penalty="l1", dual=False)
        reg_pen_dict = {'C':[10 ** i for i in range(-3,6)]}
        svm_l1_cv = GridSearchCV(estimator=svm_l1, param_grid=reg_pen_dict, cv=StratifiedKFold(5))
        clf = svm_l1.fit(X_train, y_train)
        
        y_pred = clf.predict(X_test)
        svm_test_error.append(1-accuracy_score(y_test,y_pred))
        
        #hyperplane dist determination using decision_function, which returns a confidence score for a sample
        #proportional to the signed distance of that sample to the hyperplane
        if len(active_train_df) > 0:
            hyperplane_dist = clf.decision_function(active_train_df.drop('class',axis=1))
            hyperplane_dist_df = pd.DataFrame(np.abs(hyperplane_dist))
            random_10_samples_index = hyperplane_dist_df.sort_values(by=0).iloc[:10].index
            random_10_samples = active_train_df.iloc[random_10_samples_index]
        
    mc_svm_test_error.append(svm_test_error)
```

```python
mc_active_df = pd.DataFrame(mc_svm_test_error)
```

**(c) Average the 50 test errors for each of the incrementally trained 90 SVMs in 2(b)i
and 2(b)ii. By doing so, you are performing a Monte Carlo simulation. Plot
average test error versus number of training instances for both active and passive
learners on the same figure and report your conclusions. Here, you are actually
obtaining a learning curve by Monte-Carlo simulation.**

```python
x_axis=np.arange(1,91)*10
fig, axes = plt.subplots(figsize=(12, 7))
axes.plot(x_axis, mc_passive_df.mean(), label='Passive')
axes.plot(x_axis, mc_active_df.mean(), label='Active')
axes.legend()
axes.set_xlabel("Pooled Training Data Set Size")
axes.set_ylabel("Average Test Error")
axes.set_title("Learning Curve for M=50 Monte-Carlo Simulation")
plt.show()
plt.close()
```

Clearly, we see that active learning reaches the minimum test error much faster than passive learning, reaching a minimum in around 100 observed pooled instances, compared to passive learning which needed almost all 900 pooled training data points to achieve the same results. (Rerunning gives slightly different shapes, but generally this trend prevails)
